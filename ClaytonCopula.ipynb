```python
# 2-dimensional case for the Clayton copula
cp = cop.MultiDimensionalClayton(1.5)
plt.plot(cp[0], cp[1], 'x')
```

```python
# 3-dimensional case for the Clayton copula
cp_ = cop.MultiDimensionalClayton(1.5, d=(3,1000))
```

```python
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter3D(cp_[0], cp_[1], cp_[2], cmap='hsv');
plt.show()
```

```python
data_clayton = cop.generate_data(cp_)
data_clayton
```

```python
gumb = cop.MultiDimensionalGumbel(8.5, (2,5000))
plt.plot(gumb[0], gumb[1], 'x')
```

```python
%matplotlib inline
gumb_ = cop.MultiDimensionalGumbel(8.5, (3,5000))
fig_ = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter3D(gumb_[0], gumb_[1], gumb_[2], cmap='hsv');
plt.show()
```

```python
# "2-dimensional case for the Gaussian copula"
cov = np.array([[1, 0.5], [0.5, 1]])
gauss = cop.Gaussian(cov)
plt.plot(gauss[0], gauss[1], 'x')
```
