```
from scipy.stats import multivariate_normal, norm, poisson, gamma
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
%load_ext autoreload
%autoreload 2
from CopulaGenerator import CopulaGenerator 
```

```
cop = CopulaGenerator()
```

```
# 2-dimensional case for the Clayton copula
cp = cop.MultiDimensionalClayton(7.5)
plt.plot(cp[0], cp[1], 'x')
```

```
# 3-dimensional case for the Clayton copula
cp_ = cop.MultiDimensionalClayton(7.5, d=(3,1000))
```

```
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.scatter3D(cp_[0], cp_[1], cp_[2], cmap='hsv');
plt.show()
```

```
data_clayton = cop.generate_data(cp_)
data_clayton
```

```
# "2-dimensional case for the Gaussian copula"
cov = np.array([[1, 0.5], [0.5, 1]])
gauss = cop.Gaussian(cov)
plt.plot(gauss[0], gauss[1], 'x')
```
