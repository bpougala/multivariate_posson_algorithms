```python
from scipy.stats import multivariate_normal, norm, poisson
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.random.uniform(size=50)
v = np.random.uniform(size=50)
theta = 12
cov = (u**-theta + v**-theta - 1)**-1/theta
plt.scatter(x=u, y=cov)
```

```python
u = np.random.uniform(size=5000)
v = np.random.uniform(size=5000)
theta = 12 
u_ = (u**-theta-1)**-1/theta
v_ = (v**-theta-1)**-1/theta
plt.scatter(x=u_, y=v_)
```

```python
def clayton_generator(arr, theta):
    return (arr + 1)**-1/theta

def inverse_clayton_generator(arr, theta):
    temp_clay = clayton_generator(arr, theta)
    return temp_clay**-theta - 1
```

```python
arr = np.random.randn(500)**2
theta = 8e-15
ppf_x = inverse_clayton_generator(arr, theta)
arr2 = np.random.randn(500)**2
ppf_y = inverse_clayton_generator(arr2, theta)
plt.plot(ppf_x, ppf_y, 'x')
```
